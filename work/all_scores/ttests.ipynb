```python
import pickle
import numpy as np
from scipy.stats import ttest_ind
seeds = [123, 57, 872, 340, 77, 583, 101, 178, 938, 555]
```

```python
pkl_file = open('INDEP_all_scores.pkl', 'rb')
INDEP_LSTM = pickle.load(pkl_file)
pkl_file.close()
```

```python
THETA = 54
INDEP_LSTM

INDEP_per_fold_LSTM = {}

all_metrics = ["RMSE", "RMSEat%s"%THETA]
for m in all_metrics:
    INDEP_per_fold_LSTM[m] = []
    for d in ['test']:#['train', 'valid', 'test']:
        for f in range( len(INDEP_LSTM[0][d][m]) ):
            scores=[]
            for i in range( len(seeds) ):
                scores.append( INDEP_LSTM[i][d][m][f] ) #mean over folds
            INDEP_per_fold_LSTM[m].append(scores)
            mean = np.nanmean(scores)
            std  = np.nanstd(scores)
            #if d == 'test':
            print(m, d, f, mean, std)
            print(scores)
INDEP_per_fold_LSTM
```

```python
pkl_file = open('INDEP_all_scoresQ.pkl', 'rb')
INDEP_Q = pickle.load(pkl_file)
pkl_file.close()
```

```python
THETA = 54
INDEP_Q

INDEP_per_fold_Q = {}

all_metrics = ["RMSE", "RMSEat%s"%THETA]
for m in all_metrics:
    INDEP_per_fold_Q[m] = []
    for d in ['test']:#['train', 'valid', 'test']:
        for f in range( len(INDEP_Q[0][d][m]) ):
            scores=[]
            for i in range( len(seeds) ):
                scores.append( INDEP_Q[i][d][m][f] ) #mean over folds
            INDEP_per_fold_Q[m].append(scores)
            mean = np.nanmean(scores)
            std  = np.nanstd(scores)
            #if d == 'test':
            print(m, d, f, mean, std)
            print(scores)
INDEP_per_fold_Q
```

```python
#INDEP pvalues
THETA=54
for m in ["RMSE", "RMSEat%s"%THETA ]:
    print(m)
    for f in range(len(INDEP_per_fold_Q[m])):
        data1 = INDEP_per_fold_LSTM[m][f]
        data2 = INDEP_per_fold_Q[m][f]
        statistic, pvalue = ttest_ind(data1,data2, equal_var=True)
        print(statistic, pvalue)
```
